```python
import numpy as np
import math
import matplotlib.pyplot as plt
from numpy import linalg as LA 
import functools
import fractions
import pandas as pd
```

```python
def F1(x):
    return (math.atan(x*x) - 0.4)/x
```

```python
def F2(x):
    return math.sqrt((1-0.6*x*x)/2.0)
```

```python
def G1(x, y):
    return math.sqrt(math.tan(x*y+0.4))
```

```python
def G2(x):
    return math.sqrt((1-0.6*x*x)/2.0)
```

```python
def G3(x, y):
    return (math.atan(x*x) - 0.4)/y
```

```python
def H2(x):
    return (math.atan(x*x) - 0.4)/x
```

```python
def H1(y):
    return -1.0*math.sqrt((1-2*y*y)/0.6)
```

```python
def Jacobi(x, y):
    X = np.zeros((2,2), dtype = float)
    X[0][1] = x /(2*math.sqrt(math.tan(x*y + 0.4))*(math.pow(math.cos(x*y+0.4), 2)))
    X[0][0] = y /(2*math.sqrt(math.tan(x*y + 0.4))*(math.pow(math.cos(x*y+0.4), 2)))
    X[1][0] = - 1.2*x/(2*math.sqrt(2)*math.sqrt(1 - 0.6*math.pow(x, 2)))
    return X
```

```python
def Jacobi1(x, y):
    X = np.zeros((2,2), dtype = float)
    X[0][1] = -(math.atan(x*x) - 0.4) /(math.pow(y, 2))
    X[0][0] = 2*x/(y*(1 + math.pow(x, 4)))
    X[1][0] = - 1.2*x/(2*math.sqrt(2)*math.sqrt(1 - 0.6*math.pow(x, 2)))
    return X
```

```python
# First quadrant
N = 300
X = np.zeros(N, dtype = float)
Y = np.zeros(N, dtype = float)
Z = np.zeros(N, dtype = float)
for i in range (N):
    X[i] = (250.0+i)/500.0 
    Y[i] = F1(X[i])
    Z[i] = F2(X[i])
fig = plt.figure()
plt.plot(X,Y)
plt.plot(X,Z)
plt.show()
```

```python
x_0 = 1.048
y_0 = 0.413
print((Jacobi(x_0, y_0)))
print(LA.eigvals(Jacobi(x_0, y_0)))
x = x_0
y = y_0
n_Count = 10
for i in range(n_Count):
    xHelp = x
    yHelp = y
    x = G1(xHelp, yHelp)
    y = G2(xHelp)
    print(x, y)
```
