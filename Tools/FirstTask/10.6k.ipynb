```python
import numpy as np
from numpy import linalg as LA
import math
```

```python
N = 10
A = np.zeros((N,N))
for i in range(N):
    for j in range(N):
        if (i != j):
            A[i][j] = (1)/(i+1+j+1)
        else:
            A[i][j] = 1
```

```python
print(A)
```

```python
F = np.array([1/(i+1) for i in range(N)])
```

```python
print(F)
```

```python
def yakob(A,X,F,N):
    Y = np.zeros(N)
    for i in range(N):
        for j in range(N):
            if j == i:
                continue
            Y[i] += -(X[j]*(A[i][j]))
        Y[i] += F[i]
        Y[i] = Y[i]/(A[i][i])
    X = Y
    return X
```

```python
X = np.zeros(N)
for i in range(100):
    X = yakob(A,X,F,N)
    print(X)
```

```python
print(F - A@X)
```

```python
def zeid(A,X,F,N):
    for i in range(N):
        X[i] = 0
        for j in range(N):
            if j == i:
                continue
            X[i] += -(X[j]*(A[i][j]))
        X[i] += F[i]
        X[i] = X[i]/(A[i][i])
    return X
```

```python
X = np.zeros(N)
for i in range(8):
    X = zeid(A,X,F,N)
    print(X)
```

```python
print(F - A@X)
```

```python
#Метод Гаусса
    #Прямой ход метода Гаусса
def gauss_1(A,F,N):
    C = A.copy()
    L = F.copy()
    for i in range(N):
        for j in range(i+1,N):
            coef = C[j][i]/C[i][i]
            L[j] -= L[i]*coef
            for k in range(i,N):
                C[j][k] -= C[i][k]*coef
    return C, L
```

```python
B,L = gauss_1(A,F,N) #применение прямого хода
```

```python
#Обратный ход метода Гаусса
def gauss_2(B,L,N):
    D = B.copy()
    K = L.copy()
    for i in reversed(range(N)):
        for j in range(0,i):
            coef = D[j][i]/D[i][i]
            K[j] -= K[i]*coef
            for k in range(0,i+1):
                D[j][k] -= D[i][k]*coef
    return D, K
```

```python
#Применение обратного хода метода Гаусса

R, J = gauss_2(B,L,N)
Sol = np.zeros(N)
for i in range(N):
    Sol[i] = J[i]/R[i][i]
```

```python
N = 10
L = np.zeros((N,N))
D = np.zeros((N,N))
U = np.zeros((N,N))
for i in range(N):
    for j in range(N):
        if (i == j):
            D[i][j] = A[i][j]
        else:
            if (i > j):
                L[i][j] = A[i][j]
            else:
                U[i][j] = A[i][j]
```

```python
print(L)
print(D)
print(U)
```

```python
R = -LA.inv(L+D)@U
```

```python
C = R@R.transpose()
u= F
for j in range(100):
    u = C@u/LA.norm(u, 2)
    
print((scal(u,C@u)/scal(u,u)))
norm = math.sqrt((scal(u,C@u)/scal(u,u)))
print("Евклидова норма:",norm)
```
