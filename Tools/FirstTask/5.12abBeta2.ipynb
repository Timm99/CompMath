```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x_data = np.arange(1910,2001,10)
y_data = np.array([92228496, 106021537, 123202624, 132164569, 151325798, 179323175, 203211926, 226545805, 248709873, 281421906])
N = np.size(x_data)
```

```python
#Пункт 1
```

```python
B = np.polynomial.polynomial.polyval(x_data,np.eye(2))
A = B@B.transpose()
F = B@y_data
print(B)
```

```python
c_1 = np.linalg.solve(A,F)
x_plot = np.arange(1910,2001,0.1)
y_plot = np.polynomial.polynomial.polyval(x_plot,c_1)
#Предсказание 
#Предсказание 
Predict_A_2 = np.polynomial.polynomial.polyval(2010,c_1)
print(Predict_A_2)
```

```python
fig = plt.figure()
plt.plot(x_data,y_data,"o")
plt.plot(x_plot,y_plot)
plt.show()
```

```python
z_data = x_data - 1955
```

```python
B = np.polynomial.polynomial.polyval(z_data,np.eye(2))
A = B@B.transpose()
F = B@y_data
```
