Составить таблицу с шагом h = 0.01 функции $\beta = \beta (\alpha)$, которая задана следующим образом:
$$\ddot{x} + x^3 = 1 - t^2 \hspace{10mm} 0 < t < T $$
$$x(0) = \alpha$$ 
$$\dot{x} = 0$$ 
$$-1 < \alpha < 0$$
$$\beta = x(t, \alpha)|_{t = 1} $$
Сведем это ОДУ второго порядка к системе 2 на 2:
$$\dot{x} = y$$
$$\dot{y} = -x^3 + 1 - t^2$$
$$x(0) = \alpha, y(0) = 0$$

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
Alpha = np.arange(-0.99, 0.00, 0.01)
T = 1
N = 1000
Step = 1/N 
X = np.zeros((len(Alpha), N+1), dtype = float)
Y = np.zeros((len(Alpha), N+1), dtype = float)
for i in range(len(Alpha)):
    X[i][0] = Alpha[i]
    Y[i][0] = 0
for i in range(len(Alpha)):
    for j in range(N):
        X[i][j+1] = X[i][j] + Step*Y[i][j]
        Y[i][j+1] = Y[i][j] + Step*(-((X[i][j])**3) + 1 - (j/N)**2)
```

```python
T = np.zeros(N+1)
for i in range(len(T)):
    T[i] = i/N
fig = plt.figure()
plt.plot(T, X[0])
plt.plot(T, X[1])
plt.plot(T, X[2])
plt.plot(T, X[3])
plt.plot(T, X[4])
plt.show()
```

```python
fig = plt.figure()
for i in range(len(Alpha)):
    plt.plot(T, X[i])
plt.show()
```
